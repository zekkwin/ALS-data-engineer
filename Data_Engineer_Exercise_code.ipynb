```python
# Import the appropriate packages in order read the csv files and to create dataframes.

import pandas as pd
import csv
```

```python
# Read into the three csv files.

cons = pd.read_csv('cons.csv')
cons_email = pd.read_csv('cons_email.csv')
cons_subs = pd.read_csv('cons_email_chapter_subscription.csv')
```

```python
# Sort cons.csv by one of its few columns with unique entries, cons_id.

cons_sort = cons.sort_values('cons_id')
cons_sort.head()
```

```python
# Same done with cons.csv, with the cons_email_id column instead.

cons_email_sort = cons_email.sort_values(by=['cons_email_id'])
cons_email_sort.head()
```

```python
# Same with cons_email_chapter_subscription_id.

cons_subs_sort = cons_subs.sort_values('cons_email_chapter_subscription_id')
cons_subs_sort.head()
```

```python
# Merge the sorted cons_email and cons_subs dataframes with each other.
# Note that the dataframes have different number of entries, so many rows will have NaN values.

merged_email_sub = pd.merge_ordered(cons_email_sort, cons_subs_sort, 
                                    left_on='cons_email_id', right_on='cons_email_chapter_subscription_id')
merged_email_sub.head()
```

```python
# Merge the above result the with last remaining dataframe.
# Same issue as above occurs.
final_merge = pd.merge_ordered(merged_email_sub, cons_sort, left_on='cons_email_id_x', right_on='cons_id')
final_merge.loc[0]
```

```python
# Select the necessary columns for the final output and rename them to the appropriate labels.

output = final_merge[['email', 'source', 'isunsub', 'create_dt_y', 'modified_dt_y']]
people = output.rename(columns={"create_dt_y": "created_dt", "modified_dt_y": "updated_dt", "isunsub": "is_unsub"})
people.head()
```

```python
# Save final result as a csv file.

people.to_csv('people.csv', header=['email', 'source', 'is_unsub', 'created_dt', 'updated_dt'])
```

```python
# Group the above output by its size, with the first column being the date created
# and the second column being the number of users created on that date.

peopleseries = people.groupby('created_dt').size()
acquisition_facts = peopleseries.to_frame(name = 'size').reset_index().rename(columns=
                        {'created_dt': 'acquisition_date', 'size': 'acquisitions'})
acquisition_facts.head()
```

```python
# Save the final acquisitions output as a csv file.

acquisition_facts.to_csv('acquisition_date.csv', header=['acquisition_date', 'acquisitions'])
```
